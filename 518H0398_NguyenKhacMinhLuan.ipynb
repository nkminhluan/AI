```python
import time 
k = 1
def allSolution(board):                                  ### Function numbered and print all solution
    global k 
    print(k, ":\n") 
    k = k + 1
    for i in range(8): 
        for j in range(8): 
            print(board[i][j], end = " ") 
        print("\n") 
    print("\n") 

def check(board,row,column) : 
    i,j = row,column
    while i >= 0 and j >= 0:                            ### Check if there are two queens in the upper diagonal then return false
        if(board[i][j])=='Q': 
            return False; 
        i -= 1
        j -= 1
    i,j = row,column
    while j >= 0 and i < 8:                           ### Check if there are two queens in the lower diagonal then return false 
        if(board[i][j])=='Q': 
            return False
        i += 1
        j -= 1
    for i in range(column):                           ### Check if there are two queens in the row
        if (board[row][i])=='Q':
            return False
    return True

def backtracking2(board, column) :                    ## Main function use recursive and backtracking
    if (column == 8):                                 ## If 8 queens are placed successfully then print all the solution
        allSolution(board) 
        return True
    for i in range(8): 
        if (check(board, i, column)):                 ## Check if the two queens can attack each other or not
            board[i][column] = 'Q'                     ### Place this queen in board[i][column]
            backtracking2(board, column + 1)           ### Recursion the next columnn
            board[i][column] = 0                         # Backtrack and remove queen from board[i][column]
start_time=time.time()
board = [[0 for i in range(8)] for j in range(8)]        ### Create the chessboard 
backtracking2(board,0)
end_time=time.time()
print('Total time : %fs' % ((end_time - start_time) ))
```

```python
import time
def swap(arr,start,i):
    arr[start],arr[i] = arr[i],arr[start]  ### Create function swap arr[i] and arr[start]
def permutation(arr, start, end):
    if (start == end):                     ### if arr have one element then return arr
        print(arr)
    else:
        for i in range(start, end+1 ):                     ### start from start to end + 1
            swap(arr,start,i)                             ### swap arr[i] and arr[start]
            permutation(arr, start + 1, end)   ### call recursion
            swap(arr,start,i) ## Backtracking 
A = [0,1,2]
start_time=time.time()
permutation(A, 0, len(A)-1)   ## The first index of a list is zero
end_time=time.time()
print('Total time : %fs' % ((end_time - start_time)))
```

```python
import time
board = [
    [0,0,8,0,0,0,0,0,0],
    [4,9,0,1,5,7,0,0,2],
    [0,0,3,0,0,4,1,9,0],
    [1,8,5,0,6,0,0,2,0],
    [0,0,0,0,2,0,0,6,0],
    [9,6,0,4,0,5,3,0,0],
    [0,3,0,0,7,2,0,0,4],
    [0,4,9,0,3,0,0,5,7],
    [8,2,7,0,0,9,0,1,3]
]
def print_board(bo):### function divide the board into grid 3x3
    for i in range(len(bo)):
        if i % 3 == 0 and i != 0:
            print("- - - - - - - - - - - - - -")
        for j in range(len(bo[0])):
            if j % 3 == 0 and j != 0:
                print(" . ", end =" ")
            if j == 8:
                print(bo[i][j])
            else:
                print(str(bo[i][j]) + " ", end="")
print("Sudoku board: ")
print_board(board)

def check(row,column,n):                                   ## Function to check if a n can be placed in the given block
    global board
    for i in range(9):
        if board[row][i] == n and i!=column:               ## Check row column i position if have the same element then return false
            return False
    for i in range(9):
        if board[i][column] == n and i!=row:               ## Check column row i position if have the same element then return false
            return False
    row_check = (row//3)*3                                 ##
    column_check = (column//3)*3                           ##
    for i in range(3):                                     ### FUNCTION CHECK SAME ELEMENT IN THE GRID 3x3
        for j in range(3):
            if board[row_check+i][column_check+j] == n:
                return False    
    return True
print(check(0,5,3))

def backtracking():                                         ### Function difine backtracking and input the number in the board
    global board
    for row in range(9):
        for column in range(9):
            if board[row][column] == 0:
                for n in range(1,10):                       ## input n in the board 
                    if check(row,column,n):
                        board[row][column] = n              ## Start input number
                        backtracking()
                        board[row][column] = 0              ## Backtracking will try a other number
                return
    print('\nThe Solution: ')
    start_time=time.time()
    print_board(board)
    end_time=time.time()
    print('\nTotal time : %fs' % ((end_time - start_time) ))
backtracking()
```
