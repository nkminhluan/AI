```python
def swap(arr,start,i):
    arr[start],arr[i] = arr[i],arr[start]  ### Create function swap arr[i] and arr[start]
def permutation(arr, start, end):
    if (start == end):                     ### if arr have one element then return arr
        print(arr)
    else:
        for i in range(start, end+1 ):                     ### start from start to end + 1
            swap(arr,start,i)                             ### swap arr[i] and arr[start]
            permutation(arr, start + 1, end)   ### call recursion
            swap(arr,start,i) ## Backtracking 
A = [4,5,6]
permutation(A, 0, len(A)-1)  # The first index of a list is zero
```

```python
# For each cell, we've to satisfy the following conditions:
# 1> Entry must be between integers 1 - 9.
# 2> The integer shouldn't match any other cell in it's column.
# 3> The integer shouldn't match any other cell in it's row.
# 4> The integer shouldn't match any other cell in 3x3 grid
board = [
    [0,0,8,0,0,0,0,0,0],
    [4,9,0,1,5,7,0,0,2],
    [0,0,3,0,0,4,1,9,0],
    [1,8,5,0,6,0,0,2,0],
    [0,0,0,0,2,0,0,6,0],
    [9,6,0,4,0,5,3,0,0],
    [0,3,0,0,7,2,0,0,4],
    [0,4,9,0,3,0,0,5,7],
    [8,2,7,0,0,9,0,1,3]
]
def print_board(bo):### function divide the board into grid 3x3
    for i in range(len(bo)):
        if i % 3 == 0 and i != 0:
            print("- - - - - - - - - - - - - ")
        for j in range(len(bo[0])):
            if j % 3 == 0 and j != 0:
                print(" | ", end="")
            if j == 8:
                print(bo[i][j])
            else:
                print(str(bo[i][j]) + " ", end="")
print("Sudoku board: ")
print_board(board)

#Function to check if a n can be placed in the given block
def check(row,column,n):
    global board
    for i in range(9):
        if board[row][i] == n and i!=column: ## Check row column i position if have the same element then return false
            return False
    for i in range(9):
        if board[i][column] == n and i!=row: ## Check column row i position if have the same element then return false
            return False
    row_check = (row//3)*3                                 ##
    column_check = (column//3)*3                        ##
    for i in range(3):                                     ### FUNCTION CHECK TO THE GRID 3x3
        for j in range(3):
            if board[row_check+i][column_check+j] == n:
                return False    
    return True
print(check(0,5,3))
def solve():  ### Function difine backtracking and input the number in the board
    global board
    for row in range(9):
        for column in range(9):
            if board[row][column] == 0:
                for n in range(1,10):  ## input n in the board 
                    if check(row,column,n):
                        board[row][column] = n
                        solve()
                        board[row][column] = 0  #Backtrack step
                return
    print('\nThe Solution')
    print_board(board)
solve()
```
